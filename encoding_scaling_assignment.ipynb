```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder, OrdinalEncoder
from sklearn.preprocessing import RobustScaler

# Original Dataset
data = {
    'Gender': ['Male', 'Female', 'Female', 'Male', 'Female'],
    'City': ['Mumbai', 'Delhi', 'Chennai', 'Mumbai', 'Delhi'],
    'Size': ['Small', 'Large', 'Medium', 'Large', 'Small'],
    'Age': [25, 45, 32, 60, 28],
    'Fare': [15, 300, 85, 450, 20]
}

df = pd.DataFrame(data)
print("Original DataFrame:")
display(df)
```

```python
# ====================== Task 1: Categorical Encoding ======================

# 1. Gender → Label Encoding
le_gender = LabelEncoder()
df['Gender_encoded'] = le_gender.fit_transform(df['Gender'])

# 2. City → One-Hot Encoding (drop_first=True to avoid dummy variable trap)
city_onehot = pd.get_dummies(df['City'], prefix='City', drop_first=True)
df = pd.concat([df, city_onehot], axis=1)

# 3. Size → Ordinal Encoding: Small=0, Medium=1, Large=2
size_order = [['Small', 'Medium', 'Large']]
ordinal_enc = OrdinalEncoder(categories=size_order)
df['Size_encoded'] = ordinal_enc.fit_transform(df[['Size']]).astype(int)

print("\nDataFrame after Categorical Encoding:")
display(df)
```

```python
# ====================== Task 2: Feature Scaling ======================

# We use RobustScaler because Fare has a clear outlier (450).
# RobustScaler is robust to outliers unlike StandardScaler or MinMaxScaler.

scaler = RobustScaler()

df[['Age_scaled', 'Fare_scaled']] = scaler.fit_transform(df[['Age', 'Fare']])

print("\nDataFrame after Feature Scaling:")
display(df[['Age', 'Fare', 'Age_scaled', 'Fare_scaled']])

# Final features ready for ML model
print("\nFinal Prepared Features:")
final_features = df[['Gender_encoded', 'City_Delhi', 'City_Mumbai', 'Size_encoded', 'Age_scaled', 'Fare_scaled']]
display(final_features)
```
